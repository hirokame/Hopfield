```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
import random
from tqdm import tqdm
```

```python
def data_load(path):##128*128の二値画像を返す
    path = path
    image = Image.open(path).resize((128,128),Image.LANCZOS).convert('1')
    return image
```

```python
def add_noise(image,rate): ##rate=0~1
    img=np.array(image)
    noisenum=int(128**2*rate)
    
    for i in range(noisenum):
        j=random.randint(0,127)
        k=random.randint(0,127)
        if img[j][k]==0:
            img[j][k]=255
        else:
            img[j][k]=0
    image=Image.fromarray(img)
    return image
```

```python
class hopfield:
    def __init__(self,size):
        self.len = size**2
        self.node = np.zeros(self.len)
        self.W = np.zeros((self.len, self.len))
        
    def train(self, image):
        img = np.array(image).reshape([1,-1])
        print(img.shape)
        for i in range(self.len):
            if img[0][i]>0:
                self.node[i]=1
            else:
                self.node[i]=-1
        for i in tqdm(range(self.len)):
            for j in range(self.len):
                if i==j:
                    self.W[i][j]=0
                else:
                    self.W[i][j]=self.node[i]*self.node[j]
    
    def run(self, image, step):
        img=np.array(image).reshape([1,-1])
        for i in range(self.len):
            if img[0][i]>0:
                self.node[i]=1
            else:
                self.node[i]=-1
            
        for t in tqdm(range(step)):
            i=random.randint(0,self.len-1)
            if np.dot(self.node,self.W[i])>=0:
                self.node[i]=1
            else:
                self.node[i]=-1
        
        for i in range(self.len):
            if self.node[i]>0:
                img[0][i]=255
            else:
                img[0][i]=0
        image=Image.fromarray(img.reshape([128,128]))
        return image
```

```python
image = data_load('Images/pig.png')
H = hopfield(128)
```

```python
H.train(image)
```

```python
image.save('Images/pig.jpg')
img_noise = add_noise(image,0.5)
img_noise.show()
img_noise.save('Images/noise_image.jpg')
```

```python
img_recall = H.run(img_noise,1000)
img_recall.show()
img_recall.save('Recall/recall_image1000.jpg')
```

```python
for i in range(200):
    recallnum = 1000+(i+1)*1000
    img_recall = H.run(img_recall, 1000)
    img_recall.save(f'Recall/recall_image{recallnum}.jpg')
```
